### Package NTM

```python
from NTM import *
```

### TREE data structure

```python
# create a new node
root = node('mix')

# assign its children
root.children = [node('mix2', 1), node('mix3', 3)]
root.children[0].children = [node('reagent1'), node('reagent2')]
root.children[1].children = [node('reagent3'), node('reagent4')]

print (root)
```

### View/Save TREE

```python
# view NODE data structure in tree form
viewTree(root)

# save tree at given location(image_name in folder)
saveTree(root, './folder/image_name.png')
```

### Custom User Defined TREE

```python
t = node('M1')
t.volume = 4
t.children = [node('r1'), node('M2'), node('M3')]

#child2
x = t.children[1]
x.children = [node('r1'), node('r2'), node('r2'), node('r4')]

#child3
x = t.children[2]
x.volume = 2
x.children = [node('M4'), node('M5')]

#child3_1
x = x.children[0]
x.children = [node('r1'), node('r1'), node('r2'), node('r2')]

#child3_2
x = t.children[2].children[1]
x.volume = 3
x.children = [node('r2'), node('r3'), node('r3'), node('r4')]

input_tree = t
viewTree(input_tree)
```

### listToTree TREE

```python
root = [4, 
        [2, [2, 'a', 'b', 'c', 'd'],[2, 'sa', 'b', 'c', 'd']], 
        [1, [1,'aa', 'ab', 'ac', 'ad'], [3, 'a', 'b', 'dc', 'd']],
        'r'
       ]
output_tree = listToTree(root)

viewTree(output_tree)
```

### listToSkew TREE

```python
root = [[1, 2], [11], [21,22,23], [61, 62, 63, 64]]
output_tree = listToSkewTree(root)

viewTree(output_tree)
```

### genMix TREE

```python
# input for 'genMix'
input_ratio = [('r0', 1), ('r1', 1), ('r2', 3), ('r3', 11)]
output_tree = genMix(input_ratio, 4)

viewTree(output_tree)
```

### Heuristic HDA

```python
output_tree = hda(output_tree)

viewTree(output_tree)
```

### Placement NTM

```python
output_assignment_set = ntm(output_tree, [0,0], [1])

for item in output_assignment_set:
    print(item)
```

### Visualise Placement

```python
visualize_placement(output_assignment_set, axis=[[-3,2], [-3,2]], save_dir='./11testttt')
```
